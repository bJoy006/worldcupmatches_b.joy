** Data cleaning**

```python
#print Null values
df.isnull()
```

```python
# print sum of all null values
df.isnull().sum()
```

```python
# fill Null values with the help of mean() function
df["MatchID"].fillna(df["MatchID"].mean(),inplace=True)
df.isnull().sum()
```

```python
# tring to concatenate between the str and int value with the help of condition checking
if df["Datetime"].dtype == 'object':
    df["Datetime"] = pd.to_datetime(df["Datetime"], errors='coerce')
df["Datetime"].fillna(df["Datetime"], inplace=True)
df.isnull().sum()
```
